```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_points=[]
y_points=[]

training_data=[]

with open("test.csv", "r") as file:
    content = csv.reader(file)
    next(content)
    for line in content:
        x,y = line
        x= float(x)
        y= float(y)
        x_points.append(x)
        y_points.append(y)
        training_data.append((x,y))
# print(training_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hypothesis function
def h(x, a, b):
    return a*x + b

# Initializing parameters
a = 0.0
b = 0.0

# Hyperparameters
alpha = 0.0000001   # learning rate      # ideal value 0.000001 and demo value is 0.0000001
epochs = 50     # number of epochs

# Training loop
for epoch in range(epochs):
    sum_a = 0.0
    sum_b = 0.0

    for x, y in training_data:
        y_pred = h(x, a, b)
        error = y - y_pred

        sum_a += error * x
        sum_b += error * 1

    # Gradient descent parameter update
    a += alpha * sum_a
    b += alpha * sum_b
    print(f"a = {a:.4f}, b = {b:.4f}")
    # Plot after each epoch
    plt.figure(figsize=(6, 4))
    plt.scatter(x_points, y_points, label="Data")
    x_axis = np.linspace(min(x_points), max(x_points), 200)
    plt.plot(x_axis, h(x_axis, a, b), color="red", label=f"Epoch {epoch+1}")
    plt.title(f"Logistic Regression Fit (Epoch {epoch+1})")
    plt.xlabel("x")
    plt.ylabel("y / probability")
    plt.legend()
    plt.show()

print("Final parameters:")
print(f"a = {a:.4f}, b = {b:.4f}")
```
